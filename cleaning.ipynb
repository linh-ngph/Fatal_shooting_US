```python
def gender_category(x):

    x = x.lower()
    
    # Mapping categories
    if x in ['male', 'm']:
        return 'male'
    elif x in ['female', 'f']:
        return 'female'
    elif x in ['unknown', 'undetermined', 'non-binary', 'nonbinary', 'genderqueer', 'gender fluid', 'agender']:
        return 'unknown'
    elif x in ['transgender', 'transgender woman', 'trans woman', 'trans female', 'transgender man', 'trans man', 'trans male']:
        return 'Transgender'
    else:
        return 'unknown'

# Applying the function to the DataFrame's gender column and calculating value counts
deaths['gender'] = deaths['gender'].apply(gender_category)
```

```python
deaths['race'].astype('category')
```

```python
def race_category(x):
    x = x.lower()
    
    # Mapping to the defined categories
    if x == 'white':
        return 'white_alone'
    elif x == 'black':
        return 'black_alone'
    elif x == 'hispanic':
        return 'other_race'
    elif x == 'unknown race':
        return 'other_race'
    elif x == 'asian':
        return 'asian_alone'
    elif x == 'native american':
        return 'american_indian_alaska_native_alone'
    elif x == 'pacific islander':
        return 'native_hawaiian_pacific_islander_alone'
    elif x in ['black;hispanic', 'native american;hispanic']:
        return 'other_race' 
    else:
        return 'other_race' 

# Applying the function to transform the race column
deaths['race'] = deaths['race'].apply(race_category)

```

```python
deaths['manner_of_death'].astype('category')
```

```python
def categorize_manner_of_death(x):
    x = x.lower()
    
    # Check for conditions to categorize
    if 'gunshot' in x and 'taser' in x:
        return 'shot and Tasered'
    elif 'gunshot' in x:
        return 'shot'
    else:
        return 'other'

# Applying the function to categorize the array values in the cause of death column
deaths['manner_of_death'] = deaths['manner_of_death'].apply(categorize_manner_of_death)

```

```python
def mental(x):
    x = x.strip().lower()
    
    # Mapping to True for use indications and False otherwise
    if x in ['yes', 'drug or alcohol use', 'drug or alcohol use']:
        return True
    elif x in ['no', 'no ', 'unclear', 'undetermined']:
        return False
    else:
        return 'Undetermined'  


deaths['mental_illness'] = deaths['mental_illness'].apply(mental)

```

```python
def armed_status(x):
    x = x.strip().lower()
    
    # Mapping categories
    if x == 'allegedly armed':
        return 'armed'
    elif x == 'unarmed/did not have actual weapon':
        return 'unarmed'
    elif x == 'unclear':
        return 'undetermined'
    elif x == 'vehicle':
        return 'vehicle'
    else:
        return 'undetermined'  


deaths['armed'] = deaths['armed'].apply(armed_status)
```

```python
def categorize_weapon(x):
    x = x.strip().lower()
    
    # Categorize based on keywords
    if 'gun' in x or 'firearm' in x:
        return 'gun'
    elif 'knife' in x or 'sword' in x or 'axe' in x or 'machete' in x or 'edged' in x or 'sharp' in x or 'scissors' in x or 'razor' in x:
        return 'knife'
    elif 'sword' in x or 'machete' in x or 'edged' in x or 'sharp' in x or 'scissors' in x or 'razor' in x:
        return 'sword'
    elif 'axe' in x or 'hammer' in x:
        return 'hammer'
    elif 'pole' in x or 'stick' in x or 'club' in x or 'blunt' in x or 'brick' in x or 'baton' in x or 'bat' in x:
        return 'baton'
    elif 'vehicle' in x:
        return 'vehicle'
    elif 'explosive' in x or 'grenade' in x or 'molotov' in x or 'firework' in x:
        return 'explosive'
    elif 'toy' in x or 'replica' in x:
        return 'toy Weapon'
    elif x in ['unarmed', 'no object']:
        return 'unarmed'
    elif 'unknown' in x or 'unclear' in x or 'undetermined' in x:
        return 'undetermined'
    else:
        return 'Other'

deaths['weapon'] = deaths['weapon'].apply(categorize_weapon)
```

```python
def threat_level(x):
    x = x.strip().lower()

    if x == 'attack':
        return 'attack'
    elif x in ['other', 'vehicle']:
        return 'other'
    elif x == 'undetermined':
        return 'undetermined'
    else:
        return 'undetermined'

deaths['threat_level'] = deaths['threat_level'].apply(threat_level)
```

```python
def camera(x):
    x = x.strip().lower()
    
    if x in ['yes', 'surveillance video', 'police helicopter video', 
             'dash cam video', 'bystander video', 'body cam, bystander video', 
             'body cam, dash cam video']:
        return True
    elif x in ['no', 'undetermined']:
        return False
    else:
        return False  

deaths['body_camera'] = deaths['body_camera'].apply(camera)

```

```python

def filter_dates(df, date_column, start_year=2019, end_year=2023):
    df[date_column] = pd.to_datetime(df[date_column], errors='coerce')
    
    # Define the date range
    start_date = pd.Timestamp(f"{start_year}-01-01")
    end_date = pd.Timestamp(f"{end_year}-12-31")
    
    # Filter rows within the specified date range
    filtered_df = df[(df[date_column] >= start_date) & (df[date_column] <= end_date)]
    
    return filtered_df

filtered_deaths = filter_dates(deaths, 'date')
```

**Race distribution**

```python
#Rename column
rename_race = {
    'B02001_002E':'white_alone' ,
    'B02001_003E': 'black_alone',
    'B02001_004E': 'american_indian_alaska_native_alone',
    'B02001_005E': 'asian_alone',
    'B02001_006E': 'native_hawaiian_pacific_islander_alone',
    'B02001_007E': 'some_other',
    'B02001_008E': 'two_or_more_race'
}

race.rename(rename_race, axis=1, inplace=True)

#Rename state
race['state'] = race['state'].map(fips_to_state)
```

```python
race.isna().sum()
```

```python
# Function to fill NaN values in each race column by the median count within each state
def fill_na_with_state_median(df, columns, groupby_col='state'):
    for col in columns:
        # Calculate median within each state for the specific column
        median_by_state = df.groupby(groupby_col)[col].transform('median')
        
        df[col].fillna(median_by_state, inplace=True)
    return df

# List of columns to apply the function to
race_columns = [
    'white_alone', 'black_alone', 'american_indian_alaska_native_alone', 
    'asian_alone', 'native_hawaiian_pacific_islander_alone', 'some_other', 
    'two_or_more_race'
]

race = fill_na_with_state_median(race, race_columns)
```

```python
#Combine 'some_other' and 'two_or_more_race' columns 
race['other_race'] = race['some_other'] + race['two_or_more_race']
race.drop(columns=['some_other', 'two_or_more_race'], inplace= True)
```

```python
race_distribution = race.groupby(['state', 'year'])[['white_alone', 'black_alone', 'american_indian_alaska_native_alone', 'asian_alone', 'native_hawaiian_pacific_islander_alone', 'other_race']].sum().reset_index()
```

```python
race_distribution = race_distribution.sort_values(by='year')
```
